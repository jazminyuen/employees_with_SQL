```python
import pandas as pd
import os
```

```python
# store filepath in a variable
csv = "dept_emp.csv"
```

```python
# read csv into df
csv_df = pd.read_csv(csv)
csv_df.head()
```

```python
# make new df and drop rows with duplicate emp_no, keep the last iteration
csv_df_copy = csv_df.drop_duplicates(subset="emp_no", keep='last')
csv_df_copy.head(61)
```

```python
# export clean df to csv for import in to pgAdmin
csv_df_copy.to_csv('dept_emp_clean.csv', index=False)
```
